```python
# pip install openpyxl
path = "C:/P_Project/1.Project/5-Project_data"
# main = pd.read_excel(path + '/국립중앙도서관 대출 데이터/NL_CO_LOAN_PUB_202311-7.xlsx')
Top_Lib = pd.read_excel(path + '/2.Genie_No_Detail5.6.22.xlsx')

# 상세정보를 크롤링하기위해 앨범코드만 추출
Album_codes_column = Top_Lib['SongId']

# 중복된 데이터 제거
Unique_Album_codes = list(set(Album_codes_column))
num_records = len(Unique_Album_codes)
# 결과 출력
print(f"데이터 수: {num_records}")
```

```python

headers = {
   'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# 엑셀 파일 생성
wb = Workbook()
ws = wb.active

# 헤더 추가
ws.append(['SongId', '장르', 'Release_date', 'Platform'])

Count = 0

for detail in Unique_Album_codes:
   response_Detail = requests.get("https://www.genie.co.kr/detail/songInfo?xgnm={0}".format(detail), headers=headers)
   URL_Detail = "https://www.genie.co.kr/detail/songInfo?xgnm={0}".format(detail)
         
   html_Detail = response_Detail.content.decode('utf-8','replace')
   soup_Detail = BeautifulSoup(html_Detail, 'html.parser',from_encoding='cp949')

   Count += 1
   print(f"Count : {Count}/780")
   print(URL_Detail)
   
   my_st_Detail = soup_Detail.select(
      '#body-content > div.song-main-infos > div.info-zone > ul'
   )
   for stock_name in my_st_Detail:
      # 장르
      Genre = stock_name.select_one('li:nth-child(4) > span.value').text.strip()

      # 데이터 추가
      for Genre_Text in zip(Genre):
         ws.append([detail, Genre,' ','Genle'])

      
      print(detail, Genre)
      # Count += 1
      # print(f"Count : {Count}, 총Count : {num_records}")
      
      
# 엑셀 파일 저장
file_path = r"C:\P_Project\1.Project\5-Project_data\2.Genie_detail_Data5.6.xlsx"
wb.save(file_path)
print("Excel 파일이 저장되었습니다:", file_path)
```
