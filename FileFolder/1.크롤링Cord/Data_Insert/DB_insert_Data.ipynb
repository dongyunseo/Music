```python
import pandas as pd
import mysql.connector

# MySQL 연결 설정
mydb = mysql.connector.connect(
    host='localhost',
    user="root",
    password="music",
    database="MusicDatabase"
)
mycursor = mydb.cursor()

# 엑셀 파일에서 데이터프레임으로 데이터 읽기
path = "C:/P_Project/1.Project/5-Project_data"
workbook = pd.read_excel(path + '/Insert_data.xlsx')

workbook = workbook.fillna('NULL')  # NaN 값을 'NULL' 문자열로 대체

# 테이블에 데이터 삽입
for index, row in workbook.iterrows():
    day = row['Day']
    rank = row['Rank']
    title = row['Title'].replace("'", "\\'")  # 작은 따옴표 이스케이프 처리
    singer = row['Singer'].replace("'", "\\'")  # 작은 따옴표 이스케이프 처리
    album_img_urls = row['Album_img_urls']
    album_code = row['Album_Code']
    release_date = row['Release_date'] if row['Release_date'] != 'NULL' else None  # NULL 처리
    genre_text = row['Genre_Text']
    platform = row['Platform']
    # view_count 값이 'NULL'인 경우에는 None으로 처리하고, 그렇지 않은 경우에는 정수형으로 변환
    view_count = row['view_count']
    click_url = row['Click_Url']

    # Release_date가 NULL이면 NULL 문자열로, 아니면 따옴표로 감싸서 삽입
    if release_date is None:
        release_date_sql = 'NULL'
    else:
        release_date_sql = f"'{release_date}'"

    # 삽입 쿼리 생성
    insert_query = f"INSERT INTO Music (Day, Ranking, Title, Singer, Album_img_urls, Album_Code, Release_date, Genre_Text, Platform, view_count, Click_Url) VALUES ('{day}', {rank}, '{title}', '{singer}', '{album_img_urls}', {album_code}, {release_date_sql}, '{genre_text}', '{platform}', {view_count}, '{click_url}');"

    print(insert_query)
    # 삽입 쿼리 실행
    mycursor.execute(insert_query)

# 변경사항을 데이터베이스에 반영
mydb.commit()

# 데이터베이스 연결 종료
mycursor.close()
mydb.close()
```
